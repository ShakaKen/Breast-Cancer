```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
#loading data
df = pd.read_csv("data.csv")
```

```python
print("First 5 rows")
df.head()
```

```python
#stats
print("Data Info")
df.info()
print("Description")
df.describe()
```

```python
df['Unnamed: 32'].isna().sum()
```

```python
#removing columns
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
# encoding target
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```

```python
# train-test split
from sklearn.model_selection import train_test_split
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=63)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
df.describe().T
```

```python
# visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# class balance
sns.countplot(x=df['diagnosis'])
plt.title('Diagnosis Distribution (0=Benign, 1=Malignant)')
plt.show()
```

```python
# heatmap
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
### grouping, selecting & filtering
# grouping by diagnosis to see mean values of features
grouped = df.groupby('diagnosis').mean()[['radius_mean','texture_mean','area_mean']]
print("Mean feature values by diagnosis:\n", grouped)
```

```python
# filtering high area tumors
high_area = df[df['area_mean'] > df['area_mean'].quantile(0.75)]
print(f"Number of high-area tumors: {len(high_area)}")
```

```python
# selecting subset of features
subset = df[['radius_mean','texture_mean','perimeter_mean','area_mean','diagnosis']]
print("Subset preview:\n", subset.head())
```

```python
# pivot table
table = pd.pivot_table(df, values='perimeter_mean', index='diagnosis', aggfunc='mean')
print("Average perimeter by diagnosis:\n", table)
```

```python
# boxplots
for feature in ['radius_mean', 'texture_mean', 'area_mean']:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='diagnosis', y=feature, data=df)
    plt.title(f'{feature} by Diagnosis')
    plt.show()
```
